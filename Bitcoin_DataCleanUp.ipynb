```python
# Dependencies and Setup
import os
import pandas as pd
import datetime as dt

# File to Load 
file_path = os.path.join("Resources","Bitcoin","bitcoin2020.csv")
file_path
```

```python
df20 = pd.read_csv(file_path)
df20.head()
```

```python
list(df20.columns)

```

```python
df20=df20.rename(columns={'Unnamed: 0':'Block'})
list(df20.columns)
```

```python
df20short=df20[['Block','Timestamp','Volume_(BTC)', 'Volume_(Currency)',
 'Weighted_Price']]
df20short
```

```python
df20short.columns
print(df20short.dtypes)
```

```python
df20short["Timestamp"].astype(str)
```

```python
df20short['Timestamp']= pd.to_datetime(df20short['Timestamp'])
df20short
```

```python
df20short.dtypes
```

```python
from datetime import datetime
noon = datetime.strptime('12:00:00', '%H:%M:%S').time()
noon


dfshortnoon= df20short.set_index('Timestamp').at_time(noon)
dfshortnoon
```

```python
dfshortnoon['Volume_(Currency)']=dfshortnoon['Volume_(Currency)'].map("${0:,.2f}".format)
dfshortnoon.head()
```

```python
dfshortnoon['Weighted_Price']=dfshortnoon['Weighted_Price'].map("${0:,.2f}".format)
dfshortnoon.head()
```

```python
dfshortnoon=dfshortnoon.reset_index()
dfshortnoon.head()
```

```python
dfshortnoon['Timestamp']=dfshortnoon['Timestamp'].dt.date
dfshortnoon
```

```python
dfshortnoon.dtypes
```

```python
dfshortnoon.to_csv('Resources/Bitcoin/rev_bitcoin2020data.csv')
```
